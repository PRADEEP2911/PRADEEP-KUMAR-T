# Matplotlib and Seaborn

### 1. Import the necessary libraries (Matplotlib and Seaborn) and load the data from the provided CSV file into a Pandas DataFrame.

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
d= pd.read_csv("E:\Pradeep_Datascience\Github_files\padas_question.csv")
```

```python
d
```

### 2. Create a scatter plot using Matplotlib to visualize the relationship between height and weight in the dataset. Label the axes and give the plot a title.

```python
x=d['Height (cm)']
y=d['Weight (kg)']
```

```python
plt.scatter(x, y, color='y')
plt.xlabel('Height')
plt.ylabel('Weight')
plt.title('Scatter Plot')
plt.show()
```

### 3. Create a histogram using Matplotlib to visualize the distribution of ages in the dataset. Label the axes and give the plot a title.

```python
a=d['Age']
```

```python
plt.hist(a,color='g',ec='black')
plt.xlabel('Age')
plt.ylabel('Average')
plt.title('Histogram')
plt.show()
```

###   4. Create a box plot using Seaborn to visualize the distribution of heights in the dataset by gender. Label the axes and give the plot a title

```python
sns.boxplot(x=d['Height (cm)'],y=d['Gender'])
plt.xlabel('Height')
plt.ylabel('Gender')
plt.title('Distribution of Height')
plt.show()
```

### 5. Create a bar chart using Seaborn to visualize the average weight of each gender in the dataset. Label the axes and give the plot a title.

```python
sns.barplot(x=d['Weight (kg)'],y=d['Gender'],ec='black')
plt.xlabel('Weight')
plt.ylabel('Gender')
plt.title('Average Weight')
plt.show()
```
