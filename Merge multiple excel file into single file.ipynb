## How to merge multiple excel file into single file and extract the required  columns

### Import the required library

```python
import pandas as pd
```

### Import and read the excel files that you need to merge

```python
df1=pd.read_excel("E:\Pradeep_Datascience\Test\student.xlsx")
```

```python
df2=pd.read_excel("E:\Pradeep_Datascience\Test\student_1.xlsx")
```

```python
df1
```

```python
df2
```

```python
# Find the same headline those files then write the required columns that you need in the fresh Excel


df3=df1[["name","class","mark"]].merge(df2[["name","gender","Male","Female"]], on="name", how="right")
```

```python
df3
```

```python
# Give the output directory that you need to extract the merged file

df3.to_excel("E:\Pradeep_Datascience\Test\Result.xlsx",index=False)
```
